```python
import os
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

os.makedirs("images", exist_ok=True)

# Figure 1: Airline Industry Life Cycle
fig, ax = plt.subplots(figsize=(10, 2.8))
ax.set_xlim(0, 4)
ax.set_ylim(0, 1.6)
ax.axis("off")

stages = ["Introduction", "Growth", "Maturity", "Decline"]
for i, stage in enumerate(stages):
    rect = Rectangle((i + 0.05, 0.7), 0.85, 0.35, fill=False, linewidth=1.5)
    ax.add_patch(rect)
    ax.text(i + 0.475, 0.875, stage, ha="center", va="center", fontsize=11)

for i in range(3):
    ax.annotate("", xy=(i + 1.0, 0.875), xytext=(i + 0.9, 0.875),
                arrowprops=dict(arrowstyle="->", lw=1.5))

ax.annotate("", xy=(2.475, 0.68), xytext=(2.475, 0.25),
            arrowprops=dict(arrowstyle="->", lw=1.5))
ax.text(2.475, 0.12, "Current Stage\nAirline Industry", ha="center", va="center", fontsize=10)

plt.tight_layout()
plt.savefig("images/figure1_lifecycle.png", dpi=300, bbox_inches="tight")
plt.close()

# Figure 2: Stakeholder Map
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.axis("off")

ax.text(5, 5, "Finnair", ha="center", va="center", fontsize=14,
        bbox=dict(boxstyle="round,pad=0.4", fc="white", ec="black", lw=1.5))

stakeholders = {
    "Customers": (8.5, 5),
    "Employees": (5, 1.5),
    "Regulators": (5, 8.5),
    "Suppliers": (1.5, 5),
    "Shareholders": (8, 8),
    "Alliance Partners": (2, 8),
}

for label, (x, y) in stakeholders.items():
    ax.text(x, y, label, ha="center", va="center", fontsize=11,
            bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="black"))
    ax.annotate("", xy=(x, y), xytext=(5, 5),
                arrowprops=dict(arrowstyle="-", lw=1.2))

plt.tight_layout()
plt.savefig("images/figure2_stakeholder_map.png", dpi=300, bbox_inches="tight")
plt.close()

# Figure 3: Diversification Matrix
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlim(0, 2)
ax.set_ylim(0, 2)
ax.axis("off")

# Outer box and grid
ax.add_patch(Rectangle((0, 0), 2, 2, fill=False, lw=1.5))
ax.plot([1, 1], [0, 2], color="black", lw=1.2)
ax.plot([0, 2], [1, 1], color="black", lw=1.2)

# Axis labels
ax.text(1, 2.12, "Markets", ha="center", fontsize=12)
ax.text(0.5, 1.92, "Existing", ha="center", fontsize=11)
ax.text(1.5, 1.92, "New", ha="center", fontsize=11)

ax.text(-0.18, 1, "Products", va="center", rotation=90, fontsize=12)
ax.text(-0.02, 1.5, "Existing", ha="right", va="center", fontsize=11)
ax.text(-0.02, 0.5, "New", ha="right", va="center", fontsize=11)

# Cell contents
ax.text(0.5, 1.5, "Passenger\nflights", ha="center", va="center", fontsize=11)
ax.text(1.5, 1.5, "Expanded\nroutes", ha="center", va="center", fontsize=11)
ax.text(0.5, 0.5, "Ancillary\nservices", ha="center", va="center", fontsize=11)
ax.text(1.5, 0.5, "Cargo and\nloyalty programs", ha="center", va="center", fontsize=11)

plt.tight_layout()
plt.savefig("images/figure3_diversification_matrix.png", dpi=300, bbox_inches="tight")
plt.close()

# Figure 4: Competitive Comparison
fig, ax = plt.subplots(figsize=(9, 4.5))
airlines = ["Finnair", "Lufthansa", "Ryanair"]
network_strength = [7, 10, 6]
cost_advantage = [5, 4, 10]

x = range(len(airlines))
width = 0.35

ax.bar([i - width/2 for i in x], network_strength, width=width, label="Network strength")
ax.bar([i + width/2 for i in x], cost_advantage, width=width, label="Cost advantage")

ax.set_xticks(list(x))
ax.set_xticklabels(airlines)
ax.set_ylim(0, 11)
ax.set_ylabel("Relative position")
ax.set_title("Competitive Comparison")
ax.legend()

plt.tight_layout()
plt.savefig("images/figure4_competitive_comparison.png", dpi=300, bbox_inches="tight")
plt.close()

print("All figures saved in the images folder.")
```

```python
import plotly.graph_objects as go

# Labels
x_labels = ["Existing Markets", "New Markets"]
y_labels = ["Existing Products", "New Products"]

# Cell text
cell_text = [
    ["Passenger flights", "Expanded routes"],
    ["Ancillary services", "Cargo and loyalty programs"]
]

# Numbers only to drive coloring
z = [
    [1, 2],
    [2, 3]
]

fig = go.Figure(
    data=go.Heatmap(
        z=z,
        x=x_labels,
        y=y_labels,
        colorscale="Viridis",
        showscale=False,
        hovertemplate=
            "Products: %{y}<br>"
            "Markets: %{x}<br>"
            "Strategy: %{text}<extra></extra>",
        text=[cell_text[0], cell_text[1]]
    )
)

# Add centered labels inside each cell
for yi, y in enumerate(y_labels):
    for xi, x in enumerate(x_labels):
        fig.add_annotation(
            x=x,
            y=y,
            text=f"<b>{cell_text[yi][xi]}</b>",
            showarrow=False,
            font=dict(size=16, color="white")
        )

fig.update_layout(
    title="Diversification Matrix",
    width=900,
    height=550,
    xaxis=dict(title="Markets", side="top"),
    yaxis=dict(title="Products"),
    margin=dict(l=80, r=40, t=90, b=60)
)

fig.show()

# Optional: export as an interactive HTML file
fig.write_html("diversification_matrix_interactive.html", include_plotlyjs="cdn")
```
